```python
num = 998

rev = (num % 10) * 100 + ((num // 10) % 10) * 10 + (num // 100)

print("Reverse =", rev)   #reverse third digit number
```

```python
num = 788

last = num % 10

first = num
while first >= 10:
    first = first // 10

print("First digit =", first)
print("Last digit =", last)

```
